```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
# Definir la función para extraer los datos
def extraer_datos_peliculas(generos, años):
    # Lista para almacenar la información extraída
    todas_peliculas_info = []
    
    # Iterar sobre cada año en el rango dado
    for año in años:
        # Iterar sobre cada género en la lista dada
        for genero in generos:
            # Construir la URL de la API
            url = f'https://dev.adalab.es/api/cinema/movies?year={año}&genre={genero}'
            
            # Hacer la llamada a la API
            respuesta = requests.get(url)
            
            # Verificar si la llamada fue exitosa
            if respuesta.status_code == 200:
                # Convertir la respuesta a formato JSON
                datos_peliculas = respuesta.json()
                
                # Iterar sobre cada película en los resultados
                for pelicula in datos_peliculas['results']:
                    title = pelicula['title']
                    id_pelicula = pelicula['id']
                    year = pelicula['year']
                    type_pelicula = pelicula['type']
                    
                    # Almacenar la información de la película en una tupla
                    todas_peliculas_info.append((title, id_pelicula, year, type_pelicula))
            else:
                print(f"Error en la llamada a la API para {genero} en {año}: {respuesta.status_code}")
    
    return todas_peliculas_info
```

```python
generos_a_extraer = ["Horror", "Action"]
años_a_extraer = range(2000, 2015)  # Rango de 15 años

# Extraer los datos
peliculas_extraidas = extraer_datos_peliculas(generos_a_extraer, años_a_extraer)

# Mostrar los datos extraídos
for pelicula in peliculas_extraidas:
    print(pelicula)
```

```python
# Convertir la lista de tuplas en un DataFrame
columnas = ['Title', 'ID', 'Year', 'Type']
df_peliculas = pd.DataFrame(peliculas_extraidas, columns=columnas)
```
